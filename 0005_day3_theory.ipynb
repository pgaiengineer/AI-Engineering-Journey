### The "For" Loop (Definite Iteration)
The for loop is used when you know exactly how many times you want to repeat something.

```python
# range(5) generates the numbers 0, 1, 2, 3, 4
for i in range(5):
    print(f"Processing sample number {i + 1}")
```

***Key Concept***: The variable __i__ (short for index) changes its value every time the loop runs. You can use it to keep track of where you are in a process.

### The "While" Loop (Indefinite Iteration)
The while loop is used when you want to keep going until a specific condition changes.

```python
countdown = 5

while countdown > 0:
    print(f"Centrifuge running....{countdown} seconds left.")
    countdown -= 1  # shorthand for countdown = countdown - 1

print("Process complete.")
```
